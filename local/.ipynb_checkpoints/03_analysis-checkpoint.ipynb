# 03 数据分析与可视化

基于 `01` 跑完保存的结果（`model_comparison.json`、`long_audio_eval.json`），
**无需重新跑模型**，直接加载数据进行分析和可视化。

可自由修改以下各个分析 cell，按 `Shift+Enter` 逐个执行。

## 0. 环境与数据加载

```python
import warnings
warnings.filterwarnings('ignore')

import os, json, re
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
matplotlib.use('Agg')
matplotlib.rcParams['font.family'] = ['-apple-system', 'PingFang SC', 'Microsoft YaHei', 'sans-serif']
matplotlib.rcParams['axes.unicode_minus'] = False
%matplotlib inline

REPORT_DIR = os.path.expanduser('~/Projects/Agent/local')
DATA_ROOT = os.path.expanduser('~/Desktop/hengdong_asr_trainset')

# 加载短音频结果
with open(os.path.join(REPORT_DIR, 'model_comparison.json'), 'r', encoding='utf-8') as f:
    compare_data = json.load(f)

# 加载长音频结果
with open(os.path.join(REPORT_DIR, 'long_audio_eval.json'), 'r', encoding='utf-8') as f:
    long_data = json.load(f)

models_info = compare_data['models']
samples = compare_data['samples']
print(f"短音频样本数: {len(samples)}")
print(f"长音频样本数: {len(long_data['models']['SV-base']['results'])}")
print(f"\n模型列表: {list(models_info.keys())}")
```

## 1. 短音频: 综合指标对比

```python
# 汇总表格
ordered = ['SV-base', 'SV-ft', 'Nano-base', 'Nano-ft']
print('=' * 70)
print('四模型综合指标对比 (短音频 / 全量验证集)')
print('=' * 70)
header = f'{"指标":<16}' + ''.join(f'{n:>12}' for n in ordered)
print(header)
print('-' * 70)

# CER
cers = [models_info[n]['cer'] for n in ordered]
print(f'{"CER":<16}' + ''.join(f'{c:>11.2%}' for c in cers))

# 微调相对提升
sv_rel = (models_info['SV-base']['cer'] - models_info['SV-ft']['cer']) / models_info['SV-base']['cer']
nano_rel = (models_info['Nano-base']['cer'] - models_info['Nano-ft']['cer']) / models_info['Nano-base']['cer']
print(f'{"微调相对提升":<16} {sv_rel:>+11.1%}          {nano_rel:>+10.1%}')

# 精确匹配率
exact_rates = [models_info[n]['exact_rate'] for n in ordered]
print(f'{"精确匹配率":<16}' + ''.join(f'{e:>11.1%}' for e in exact_rates))

# 速度
speeds = [models_info[n]['speed'] for n in ordered]
print(f'{"速度(条/s)":<16}' + ''.join(f'{s:>11.1f}' for s in speeds))

# 耗时
times = [models_info[n]['time'] for n in ordered]
print(f'{"推理耗时":<16}' + ''.join(f'{t:>10.1f}s' for t in times))
print('=' * 70)
```

## 2. 短音频: CER 分布直方图

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.flatten()

cer_fields = ['sv_base_cer', 'sv_ft_cer', 'nano_base_cer', 'nano_ft_cer']
titles = ['SV-base', 'SV-ft', 'Nano-base', 'Nano-ft']
colors = ['#e74c3c', '#27ae60', '#e67e22', '#3498db']

for ax, field, title, color in zip(axes, cer_fields, titles, colors):
    vals = [s[field] for s in samples]
    ax.hist(vals, bins=50, color=color, alpha=0.7, edgecolor='white')
    ax.axvline(np.mean(vals), color='black', linestyle='--', linewidth=1.5, label=f'Mean: {np.mean(vals):.1%}')
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xlabel('CER')
    ax.set_ylabel('Count')
    ax.legend()
    ax.set_xlim(0, 2)

plt.suptitle('短音频 CER 分布', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR, '03_cer_distribution.png'), dpi=150, bbox_inches='tight')
plt.show()
print('已保存: 03_cer_distribution.png')
```

## 3. 短音频: CER 箱线图对比

```python
fig, ax = plt.subplots(figsize=(10, 6))

cer_data = [[s['sv_base_cer'] for s in samples],
            [s['sv_ft_cer'] for s in samples],
            [s['nano_base_cer'] for s in samples],
            [s['nano_ft_cer'] for s in samples]]

bp = ax.boxplot(cer_data, labels=ordered, patch_artist=True)
colors = ['#e74c3c', '#27ae60', '#e67e22', '#3498db']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)

ax.set_ylabel('CER', fontsize=12)
ax.set_title('四模型 CER 箱线图对比', fontsize=14, fontweight='bold')
ax.set_ylim(0, 2)
ax.grid(axis='y', alpha=0.3)

# 标注均值
for i, (data, color) in enumerate(zip(cer_data, colors), 1):
    ax.scatter(i, np.mean(data), color='black', s=80, zorder=5, marker='D')
    ax.annotate(f'{np.mean(data):.1%}', (i, np.mean(data)), textcoords='offset points', 
                xytext=(15, 0), fontsize=10)

plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR, '03_cer_boxplot.png'), dpi=150, bbox_inches='tight')
plt.show()
print('已保存: 03_cer_boxplot.png')
```

## 4. 短音频: 微调提升分布

```python
# 计算每个样本的微调提升（正值 = 微调变好）
sv_improvement = [s['sv_base_cer'] - s['sv_ft_cer'] for s in samples]
nano_improvement = [s['nano_base_cer'] - s['nano_ft_cer'] for s in samples]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# SV 微调提升
axes[0].hist(sv_improvement, bins=50, color='#27ae60', alpha=0.7, edgecolor='white')
axes[0].axvline(0, color='red', linestyle='--', linewidth=2, label='No change')
axes[0].axvline(np.mean(sv_improvement), color='black', linestyle='--', linewidth=2, label=f'Mean: {np.mean(sv_improvement):.2f}')
axes[0].set_xlabel('CER 降低值（正值=微调变好）')
axes[0].set_ylabel('Count')
axes[0].set_title('SenseVoice 微调提升分布', fontweight='bold')
axes[0].legend()

# Nano 微调提升
axes[1].hist(nano_improvement, bins=50, color='#3498db', alpha=0.7, edgecolor='white')
axes[1].axvline(0, color='red', linestyle='--', linewidth=2, label='No change')
axes[1].axvline(np.mean(nano_improvement), color='black', linestyle='--', linewidth=2, label=f'Mean: {np.mean(nano_improvement):.2f}')
axes[1].set_xlabel('CER 降低值（正值=微调变好）')
axes[1].set_ylabel('Count')
axes[1].set_title('Fun-ASR-Nano 微调提升分布', fontweight='bold')
axes[1].legend()

plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR, '03_ft_improvement.png'), dpi=150, bbox_inches='tight')
plt.show()

# 统计
sv_better = sum(1 for x in sv_improvement if x > 0)
sv_worse = sum(1 for x in sv_improvement if x < 0)
nano_better = sum(1 for x in nano_improvement if x > 0)
nano_worse = sum(1 for x in nano_improvement if x < 0)
print(f'SV-ft: {sv_better} 变好, {sv_worse} 变差, {len(samples)-sv_better-sv_worse} 不变')
print(f'Nano-ft: {nano_better} 变好, {nano_worse} 变差, {len(samples)-nano_better-nano_worse} 不变')
```

## 5. 短音频: 按文本长度分组分析

```python
# 按参考文本长度分组
def length_group(text):
    n = len(text)
    if n <= 2: return '1-2'
    if n <= 5: return '3-5'
    if n <= 10: return '6-10'
    return '11+'

for s in samples:
    s['ref_len'] = len(s['expected'])
    s['len_group'] = length_group(s['expected'])

# 分组统计
groups = ['1-2', '3-5', '6-10', '11+']
group_data = {g: {'sv_base': [], 'sv_ft': [], 'nano_base': [], 'nano_ft': []} for g in groups}

for s in samples:
    g = s['len_group']
    group_data[g]['sv_base'].append(s['sv_base_cer'])
    group_data[g]['sv_ft'].append(s['sv_ft_cer'])
    group_data[g]['nano_base'].append(s['nano_base_cer'])
    group_data[g]['nano_ft'].append(s['nano_ft_cer'])

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

model_colors = {'sv_base': '#e74c3c', 'sv_ft': '#27ae60', 'nano_base': '#e67e22', 'nano_ft': '#3498db'}

for ax, g in zip(axes, groups):
    vals = group_data[g]
    x = np.arange(len(vals))
    width = 0.2
    for i, (k, color) in enumerate(model_colors.items()):
        means = [np.mean(v) for v in vals.values()]
        ax.bar(x + i*width, means, width, label=k, color=color, alpha=0.8)
    ax.set_xlabel('Model')
    ax.set_ylabel('Mean CER')
    ax.set_title(f'文本长度 {g} 字符', fontweight='bold')
    ax.set_xticks(x + 1.5*width)
    ax.set_xticklabels(['SV-base', 'SV-ft', 'Nano-base', 'Nano-ft'], rotation=15)
    ax.legend(fontsize=8)
    ax.set_ylim(0, 1)

plt.suptitle('按参考文本长度分组的 CER', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR, '03_cer_by_length.png'), dpi=150, bbox_inches='tight')
plt.show()
print('已保存: 03_cer_by_length.png')
```

## 6. 短音频: 最难样本分析（Top 20 错误率最高）

```python
# 取四个模型中 CER 最高的样本
samples_with_max = [(s, max(s['sv_base_cer'], s['sv_ft_cer'], s['nano_base_cer'], s['nano_ft_cer'])) for s in samples]
samples_sorted = sorted(samples_with_max, key=lambda x: -x[1])

print('--- 最难样本 Top 20 (CER 最高) ---')
print(f'{"ID":<5} {"参考文本":<12} {"SV-base":>8} {"SV-ft":>8} {"Nano-base":>10} {"Nano-ft":>10}')
print('-' * 60)
for s, worst in samples_sorted[:20]:
    exp = s['expected'][:10]
    print(f"{s['id']:<5} {exp:<12} {s['sv_base_cer']:>7.0%} {s['sv_ft_cer']:>7.0%} {s['nano_base_cer']:>9.0%} {s['nano_ft_cer']:>9.0%}")
```

## 7. 短音频: 模型预测一致性分析

```python
# 分析四个模型预测结果的一致性
def all_same(preds):
    return len(set(preds)) == 1

def all_correct(s):
    return all(s[f'{m}_cer'] == 0 for m in ['sv_base', 'sv_ft', 'nano_base', 'nano_ft'])

def all_wrong(s):
    return all(s[f'{m}_cer'] > 0 for m in ['sv_base', 'sv_ft', 'nano_base', 'nano_ft'])

def majority_correct(s):
    correct_count = sum(1 for m in ['sv_base', 'sv_ft', 'nano_base', 'nano_ft'] if s[f'{m}_cer'] == 0)
    return 0 < correct_count < 4

n_all_correct = sum(1 for s in samples if all_correct(s))
n_all_wrong = sum(1 for s in samples if all_wrong(s))
n_majority = sum(1 for s in samples if majority_correct(s))

fig, ax = plt.subplots(figsize=(8, 6))

labels = ['All Correct', 'Majority Correct', 'All Wrong']
sizes = [n_all_correct, n_majority, n_all_wrong]
colors = ['#27ae60', '#f39c12', '#e74c3c']
explode = (0.05, 0, 0.05)

wedges, texts, autotexts = ax.pie(sizes, explode=explode, labels=labels, colors=colors,
                                   autopct='%1.1f%%', startangle=90, textprops={'fontsize': 11})
ax.set_title('四模型预测一致性分布', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR, '03_model_agreement.png'), dpi=150, bbox_inches='tight')
plt.show()

print(f'四模型全部正确: {n_all_correct} ({n_all_correct/len(samples):.1%})')
print(f'四模型全部错误: {n_all_wrong} ({n_all_wrong/len(samples):.1%})')
print(f'多数正确: {n_majority} ({n_majority/len(samples):.1%})')
```

## 8. 长音频: 四模型 CER 对比

```python
long_ordered = ['SV-base', 'SV-ft', 'Nano-base', 'Nano-ft']
long_models = long_data['models']

fig, ax = plt.subplots(figsize=(10, 6))

cers = [long_models[n]['avg_cer'] for n in long_ordered]
bars = ax.bar(long_ordered, cers, color=['#e74c3c', '#27ae60', '#e67e22', '#3498db'], alpha=0.8)

ax.set_ylabel('平均 CER', fontsize=12)
ax.set_title('长音频四模型 CER 对比', fontsize=14, fontweight='bold')
ax.set_ylim(0, max(cers) * 1.2)
ax.grid(axis='y', alpha=0.3)

# 标注数值
for bar, cer in zip(bars, cers):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 10,
            f'{cer:.1%}', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR, '03_long_cer.png'), dpi=150, bbox_inches='tight')
plt.show()

# 详细数据
print('\n长音频详细数据:')
for n in long_ordered:
    r = long_models[n]
    print(f"  {n}: CER={r['avg_cer']:.1%}, samples={r['total']}, time={r['time']:.1f}s")
```

## 9. 长音频: 分 speaker 对比

```python
# 按 speaker 分组
speaker_results = {}
for n in long_ordered:
    for r in long_models[n]['results']:
        sp = r['speaker']
        if sp not in speaker_results:
            speaker_results[sp] = {}
        speaker_results[sp][n] = r['cer']

fig, ax = plt.subplots(figsize=(12, 6))

speakers = list(speaker_results.keys())
x = np.arange(len(speakers))
width = 0.2
colors = ['#e74c3c', '#27ae60', '#e67e22', '#3498db']

for i, (n, color) in enumerate(zip(long_ordered, colors)):
    vals = [speaker_results[sp][n] for sp in speakers]
    ax.bar(x + i*width, vals, width, label=n, color=color, alpha=0.8)

ax.set_xlabel('Speaker', fontsize=12)
ax.set_ylabel('CER', fontsize=12)
ax.set_title('长音频各 Speaker CER 对比', fontsize=14, fontweight='bold')
ax.set_xticks(x + 1.5*width)
ax.set_xticklabels(speakers, rotation=15)
ax.legend()
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig(os.path.join(REPORT_DIR, '03_long_by_speaker.png'), dpi=150, bbox_inches='tight')
plt.show()
print('已保存: 03_long_by_speaker.png')
```
